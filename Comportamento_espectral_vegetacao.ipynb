```python
#@title Baixando medidas e salvando em uma pasta/ importando bibliotecas
!wget https://github.com/Liaons/comportamento_espectral/raw/main/Medidas%20Aula.zip
!mkdir medidas_lab
!unzip Medidas\ Aula.zip -d medidas_labn

import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
#@title Listando os arquivos baixados e montando o dataframe
!ls medidas_lab

txt_files = glob.glob('/content/medidas_lab/*.txt')
dfs = []

# Iterate through each file, read it, and store in the list
for file_path in txt_files:
  # Extract the filename
  file_name = file_path.split('/')[-1].replace('.asd.txt', '')
  # Read the CSV file
  df = pd.read_csv(file_path, sep='\t') # Adjust sep if needed
  # Rename the data column to the file name
  df = df.rename(columns={df.columns[1]: file_name})
  dfs.append(df)

# Unindo a tabela completa
merged_df = dfs[0]
for i in range(1, len(dfs)):
  merged_df = pd.merge(merged_df, dfs[i], on='Wavelength')

merged_df.head()
```

```python
#@title Organizando os dados, calculando a média e std da radiância e fator de reflectância.
esp1 = ["esp1_dorsal_2f", "esp1_dorsal_3f", "esp1_dorsal_ou_mais", "esp1_front_2f", "esp1_front_3f", "esp1_front_4_ou_mais"]
esp2 = ["esp2_dorsal_1f","esp2_front_1f","esp2_front_v"]
esp8 = ["esp8_dorsal_v","esp8_front_v"]
esp11 = ["esp11_dorsal_v","esp11_front_v"]
extra= ["papel_cartao","placa_ref"]

prefixos = esp1 + esp2 + esp8 + esp11 + extra
colunas_filtradas = [col for col in merged_df.columns if any(col.startswith(p) for p in prefixos)]

# Calcular a média e o desvio padrão para as colunas filtradas
df_stats = merged_df[['Wavelength']].copy()

for prefixo in prefixos:
  cols_com_prefixo = [col for col in colunas_filtradas if col.startswith(prefixo)]
  if cols_com_prefixo:
    df_stats[f'{prefixo}_mean'] = merged_df[cols_com_prefixo].mean(axis=1)
    df_stats[f'{prefixo}_std'] = merged_df[cols_com_prefixo].std(axis=1)

# Calculando a reflectância
def calcular_reflectancia(df_stats, alvos, placa='placa_ref_mean'):
  """
  Calcula a reflectância para cada comprimento de onda.
  Args:
    df_stats (pd.DataFrame): DataFrame com as estatísticas de média e desvio padrão.
    alvos (list): Lista de nomes de colunas (médias) para calcular a reflectância.
    placa (str): Nome da coluna de referência (placa_ref_mean por padrão).
  Returns:
    pd.DataFrame: DataFrame com a reflectância calculada.
  """
  df_reflec = pd.DataFrame()
  df_reflec['Wavelength'] = df_stats['Wavelength']
  for col in alvos:
    df_reflec[col.replace('_mean', '')] = df_stats[col] / df_stats[placa]
  return df_reflec

# Identifique quais colunas de média você quer calcular a reflectância
colunas_de_media = [col for col in df_stats.columns if col.endswith('_mean') and col != 'placa_ref_mean']

# Calcule a reflectância usando a função
df_reflec = calcular_reflectancia(df_stats, colunas_de_media)
df_reflec.head()
```

```python
#@title Plotando o fator de Reflectância Especie 1 - diferentes numeros de folhas
plt.figure(figsize=(12,6))

for solo, cor, label in [
    (df_reflec['esp1_dorsal_2f'],       '#e7d4e8', 'Abaxial - 2 Folhas'),
    (df_reflec['esp1_dorsal_3f'],       '#af8dc3', 'Abaxial - 3 Folhas'),
    (df_reflec['esp1_dorsal_ou_mais'],  '#762a83', 'Abaxial - 4 ou mais Folhas'),
    (df_reflec['esp1_front_2f'],        '#d9f0d3', 'Adaxial - 2 Folhas'),
    (df_reflec['esp1_front_3f'],        '#7fbf7b', 'Adaxial - 3 Folhas'),
    (df_reflec['esp1_front_4_ou_mais'], '#1b7837', 'Adaxial - 4 ou mais Folhas'),
    (df_reflec['papel_cartao'],          "black", 'Papel verde'),
    ]:

    plt.plot(df_reflec['Wavelength'], solo, color= cor, label= label)

plt.xlabel('Comprimento de Onda (nm)')
plt.ylabel('Fator de Reflectância')
plt.title('Fator de Reflectância entre diferentes composições para a espécie 1')
plt.legend()
plt.show()

#Salvando a figura
plt.savefig('Solos_dry.png', dpi=300, bbox_inches='tight')
#files.download('Solos_dry.png')
```

```python
#@title Plotando o fator de Reflectância em 4 ou mais folhas, entre diferentes espécies dorsal
plt.figure(figsize=(12,6))

for solo, cor, label in [
    (df_reflec['esp1_dorsal_ou_mais'],  '#a6cee3', 'Espécie 1 - Abaxial'),
    (df_reflec['esp8_dorsal_v'],        '#fb9a99', 'Espécie 3 - Abaxial'),
    (df_reflec['esp11_dorsal_v'],       '#cab2d6', 'Espécie 4 - Abaxial'),
    ]:

    plt.plot(df_reflec['Wavelength'], solo, color= cor, label= label)

plt.xlabel('Comprimento de Onda (nm)')
plt.ylabel('Fator de Reflectância')
plt.title('Fator de Reflectância entre diferentes composições  entre espécies (dorsal)')
plt.legend()
plt.show()

#Salvando a figura
plt.savefig('Solos_dry.png', dpi=300, bbox_inches='tight')
#files.download('Solos_dry.png')
```
